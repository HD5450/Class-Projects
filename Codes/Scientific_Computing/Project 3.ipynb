```
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```
# Model
def func(t, a, b):
  return a*np.exp(b * t)

# Initial data
xdata = np.array([1,2,3,4,5])
ydata = np.array([16,27,45,74,122])

# Curve fitting
param,ph = curve_fit(func, xdata, ydata)
print(param)
```

```
# Generate data points for model function
t = np.linspace(0, 5, 50)

# Plotting
plt.plot(xdata, ydata, 'X', label='Original data', markersize = 13, c = 'black')
plt.plot(t, param[0]*np.exp(param[1]*t), 'r', label='Fitted line')
plt.xlabel('t')
plt.ylabel('values')
plt.legend()
plt.show()
```
