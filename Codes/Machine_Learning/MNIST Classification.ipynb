```
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from sklearn.datasets import fetch_mldata
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

seed = 42
numpy.random.seed(seed)

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()


num_of_pixels = X_train.shape[1] * X_train.shape[2]
Temp_X_train = X_train.reshape(X_train.shape[0], num_of_pixels).astype('float32')
Temp_X_test = X_test.reshape(X_test.shape[0], num_of_pixels).astype('float32')
Temp_X_train = Temp_X_train / 255
Temp_X_test = Temp_X_test / 255

Temp_Y_train = np_utils.to_categorical(Y_train)
Temp_Y_test = np_utils.to_categorical(Y_test)
num_classes = Temp_Y_test.shape[1]

pca = PCA(.75)
X_train_pca = pca.inverse_transform(pca.fit_transform(Temp_X_train))

my_model = Sequential()
my_model.add(Dense(10, input_dim=num_of_pixels, activation='relu'))
my_model.add(Dense(num_classes, activation='softmax'))

my_model.compile(loss='categorical_crossentropy', optimizer='RMSprop', metrics=['accuracy'])

my_model.fit(X_train_pca, Temp_Y_train, epochs=10, batch_size=100, verbose=2)

scores = my_model.evaluate(Temp_X_test, Temp_Y_test, verbose=0)
print("Error: %.2f%%" % (100-scores[1]*100))
```
