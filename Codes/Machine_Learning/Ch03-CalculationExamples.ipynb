## Calculating Bragg's Law (Ex 3.1)

```python
# n: integer
# lambda: wavelength, specify your units
# theta = angle in degrees
# returns d with units
def d(n,lda,theta): 
    return (n*lda/(2*np.sin(np.radians(theta))))
```

Try out our function (check that it matches Eq. 3.18)

```python
d(1, 0.25*u.nm, 26.3)
```

```python
listofthetas = [14.5, 22.6, 30.0]
d(1, 0.2*u.nm, listofthetas)
```

Give the answers in nm, instead

```python
d(1, 0.2*u.nm, listofthetas)/u.nm
```

This matches the result of Example 3.1, and gives us the next few peaks as well!
