```
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
```

```
import numpy as np                   #this is our old friend, the handy NumPy
import matplotlib.pyplot as plt      # matplotlib takes cares of ploting   
fig, ax = plt.subplots()


def XOR(a,b):
  return np.where(a != b, 1, 0)

      
training_size=4      
# we create a synthetic data set      
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
y_train = np.zeros((1,4))

y_train=XOR(x_train[:,0],x_train[:,1])


print(y_train)

for xx in x_train:
  if (XOR(xx[0],xx[1])==1):
    ax.plot(xx[0],xx[1],'b.')
  else:
    ax.plot(xx[0],xx[1],'r.')

plt.xlabel("a", fontsize=25)
plt.ylabel("b", fontsize=25)
plt.title('XOR(a,b). Reds are zero, and blues are 1')
```

```
my_model = Sequential()
my_model.add(Dense(8, input_dim=2, activation='relu',use_bias=1))
my_model.add(Dense(1, activation='sigmoid',use_bias=1))
```

```

my_model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```
my_model.summary()
```

```
my_model.fit(x_train, y_train,  batch_size=4, nb_epoch=600)
```

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
import keras

training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

target_data = np.array([[0],[1],[1],[0]], "float32")

model = Sequential()
model.add(Dense(2, input_dim=2, activation='relu', kernel_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.5, seed=1),
                bias_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.5, seed=2)))

model.add(Dense(1, activation='sigmoid', kernel_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.5, seed=3),
                bias_initializer=keras.initializers.RandomNormal(mean=0.0, stddev=0.5, seed=4)))

model.compile(loss='binary_crossentropy',
              optimizer='sgd',
              metrics=['binary_accuracy'])

model.fit(training_data, target_data, nb_epoch=50, verbose=2)
```
