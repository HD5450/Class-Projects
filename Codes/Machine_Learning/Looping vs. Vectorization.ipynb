```
import numpy as np


array_size= 10000000

seed = 4
np.random.seed(seed)

my_array1=np.random.rand(array_size)
my_array2=np.random.rand(array_size)

print(my_array1)
print(my_array2)
```

my_array1 and my_array2 are two one dimensional arrays. Their dot product (also known as scalar product)  would be a single scalar. Implement this dot product on these two arrays using looping. If you don't remember what was the syntax for looping in Python that's all OK. Google it or ask your friends!

```
product = 0
for i in range (array_size):
  product += my_array1[i] * my_array2[i]
print(product)
```
